```python
### Numbers and Match in Python.
# In AI and Machine Learning, numbers are absolutely everywhere — calculating accuracy of a model, measuring errors, scaling data, computing probabilities.
# Understanding how Python handles numbers is critical for everything that comes later!
# Python can do all mathematical operations

a = 100
b = 2

# Basic mathematical operations
print("Addition:", a+b)
print("Subtraction:", a-b)
print("Multiplication:", a*b)
print("Division:", a/b)
print("Floor Division:", a//b) # Removed decimal part
print("Remainder:", a%b) # Called module, The % remainder operator is used a lot in AI code — for example checking if a number is even or odd!
print("Power:", a**b) # a to the power of b
```

```python
### Rounding and Built in match functions
import math

Pi_value = 3.141592
print("Original:", Pi_value)
print("Rounded:", round(Pi_value)) # rounds to nearest number
print("Rounded to 2 decomals:", round(Pi_value, 2))
print("Rounded to 4 decimals:", round(Pi_value, 4))

#math functions
print("\nMath Functions")
print("Square root of 144:", math.sqrt(144))
print("Pi value:", math.pi)
print("Ceiling of 3.14:", math.ceil(3.14)) #always goes UP no matter what
print("Floor of 3.14:", math.floor(3.14)) #always goes DOWN no matter what
print("Abs of -50:", abs(-50)) #removes the negative sign. This is used constantly in AI to calculate error distances between predicted and actual values! 
```

```python
### Math with Variables : Real AI engineering calculations!

#Imagine you are evaluating an AI model.
total_preductions = 1000
correct_predictions = 873
wrong_predictions = total_preductions - correct_predictions

#calculate accuracy
accuracy = (correct_predictions/total_preductions)*100
error_rate = (wrong_predictions/total_preductions)*100

print("=====AI Model Evalution=====")
print(f"Total preductions: {total_preductions}")
print(f"Correct predictions: {correct_predictions}")
print(f"Wrong predictions: {wrong_predictions}")
print(f"Accuracy: {round(accuracy, 2)}%")
print(f"Error rate: {round(error_rate, 2)}%")
print(f"Is accuracy above 80% {accuracy>80}")
```

```python
### Number types and Conversion
#Converting between int, float and string
# Type conversion
my_int = 42
my_float = 3.99
my_string = "100"

#Convert int to float
print("Int to Float:", float(my_int))
#Convert float to int( removes the decimals!)
print("Float to Int:", int(my_float))
#int(3.99) = 3 — it always chops the decimal, never rounds up.
#That's why math.ceil() exists for when you actually want to round up!
#Convert int to string 
print("Int to String:", str(my_int))
#Convert String to int
print("String to Number:", int(my_string))
print("Check the type:", type(str(my_int)))

#Practical example
user_input = "85" # user typed a number but it came as string
score = int(user_input)  + 15
#converting to int first is critical — always convert user input before doing math with it!
# If we did not convert it to int, the operation user_input +15 teated as concatenation(+) but concatenation is applicable for only strings.
print(f"\nUser score: {user_input}")
print(f"Bonus added: {score}") 
```

```python
### Real world example - AI Studay tracker calculator
#Your leaning data
daily_hours = float(input("How many hours do you study per day?"))
weeks = int(input("How many weeks do you plan to studay?"))

#Calculations
total_days = weeks*7
total_hours = daily_hours*total_days
monthly_hours = daily_hours*30

print(f"\n===== Your AI Studay Planer=====")
print(f"Daily studay hours: {daily_hours}")
print(f"Study weeks: {weeks}")
print(f"Total days: {total_days}")
print(f"Total hours: {round(total_hours), 1}")
print(f"Monthly hours: {round(monthly_hours), 1}")
print(f"On track for 240 hours {total_hours>= 240}")
```
