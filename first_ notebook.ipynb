```python
_# My first data visualization using matplotlib
import matplotlib.pyplot as plt

# My 6 months learning journey phases
phases = ["Python", "Core ML", "Deep Learning", "AI Agents", "Portfolio"]
hours = [40, 55, 65, 35, 25]

# Draw a bar chart
plt.figure(figsize=(8,4))
plt.bar(phases, hours, color=["#378ADD", "#1D9E75", "#BA7517", "#7F77DD", "#D85A30"] )
plt.title ("Siva's AI Journey - Study hours per phase")
plt.xlabel("phases")
plt.ylabel("hours")
plt.tight_layout()
plt.show
           
```
