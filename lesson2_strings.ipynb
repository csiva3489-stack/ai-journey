```python
# Strings in deepth
# Strings in plain English:A String is simply any text wrapped in quotes.
# It could be a single letter, a word, a sentence, or even an entire paragraph.
# In AI, strings are everywhere — chatbots read strings, speech recognition converts voice to strings, and language models like ChatGPT generate strings!

sentence = "I am learning AI Engineering in Hyderabad"
single_word = "python"
number_as_text = "123" #this is string not an integer

#len() tels you how many charactors in a string

print("sentance", sentence)
print("word", single_word)
print("number as text", number_as_text)
print("type of number_as_text", type(number_as_text))
print("length of sentance", len(sentence))
print("length of single_word", len(single_word) )

```

```python
# String CONCATENATION : Joining strings togather
# use + to combine strings

first_name = "Siva"
last_name = "Kumar"
city = "Hyderabad"
goal = "AI Engineering"

#Methgod 01: using + operator
fullname = first_name+" "+last_name
print(fullname)

#Method 02: using f-string : Most popular in AI engineering.
introduction = f"My name is {first_name} {last_name} i live in {city} city and i am learning {goal}!"
print(introduction)

#repeating a string
line = "-"*40
print(line)
```

```python
# String Methods : Built in tools to manipulate text
#These are heavyly used in AI and NLP

my_text = "I am becoming an AI Engineer in Hyderabad"

#Change case
print("original:", my_text)
print("upper case:", my_text.upper())
print("lower case:", my_text.lower())
print("title case:", my_text.title())

#Removing spaces
messy_text = "  top many spaces aroud me  "
print("messy text:", messy_text)
print("cleaned:", messy_text.strip())

# Replace a word
about_me = my_text.replace("Hyderabad", "India")
print("Replaced:", about_me)

#Check if a word exists inside a string
print("Contains Enginner:", "Engineer" in my_text)
print("Contains Doctor:", "Doctor" in my_text)
```

```python
## String slicing
#Think of each charecter having position number startgin from 0

my_string = "AI Engineer"
#Positions : A= 0, I = 1, ''= 2, E=3, n=4, g=5, i=6, n=7, e=8, e=9, r=10

#lets get single scharecter
print("fist charecter:", my_string[0])
print("second charecter:", my_string[1])
print("last charecter:", my_string[-1])

#lets get Range of charecter [start:end]
print("first two charecters:", my_string[0:2])
print("from charcter 3:", my_string[3:])
print("from last to first:", my_string[-1:0])
print("last four charecter:", my_string[-4:])
print("everything:", my_string[:])

#my_string[0:2] printed "AI" — it grabbed characters at position 0 and 1 but stopped before position 2.
# That's the slicing rule — [start:end] always includes start but excludes end. Very important to remember!
```

```python
## String Splitting : Breaking a string into a list of words
#This is used heavily in AI and NLP to process the text
sentence = "I Am Learning AI Engineering in Hyderabad"
words = sentence.split()

#split()  breaks a string into individual words, split() broke it perfectly at every space!
print("Origianl sentence:", sentence)
print("split into words", words)
print("length of sentence:", len(words))
print("first letter of word:", words[0])
print("last letter of word:", words[-1])

# split by a specific charecter
data = "Python, Maths, Prompt Engineering, ML, AI agents, Project"
topics = data.split(",")
print("original data:", data)
print("split by comma:", topics)
print("first topic:", topics[0])
print("last topic:", topics[-1])
print("combine all topics:", f"{topics[0]} + {topics[1]} + {topics[2]} + {topics[3]} + {topics[4]} + {topics[5]}")
```

```python
# Putting it all togather - Real world string manipulation
#This is how exactly how AI program process text

#Imagine this is a user message to an AI chatbot
user_message = "  Hello! i want to learn about Machine Learning  "

#Step 01- Clean the message
cleaned = user_message.strip()
print("Cleaned:", cleaned)

#Step 02 - Convert to lower case for processing
lowered = cleaned.lower()
print("lowered:", lowered)

#Step 03 - Check the what user want
if "machine learning" in lowered :
    print("Topic detected: Machine Learning!")

#step 04 - Count words in message
count_words = len(lowered.split())
print("Count of words:", count_words)

#Step 05 - Build a response using f-string
response = f"Hello!, you have sent a {count_words} word message about Machine Learning!"
print("AI Response:", response)
```
