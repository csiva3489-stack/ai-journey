```python
# STRINGS - Used to store test
#Always wrapped in quotes

my_name = "Siva"
my_city = "Hyderabad"
my_goal = "AI Engineer"

print("Name:", my_name)
print("City:", my_city)
print("My Goal:", my_goal)
print("type:", type(my_name))
```

```python
# INTEGERS - Whole numbers but not decimals
# used for counting, ages, years etc.

my_age = 28
study_hours_per_week = 10
months_to_goal = 6
total_weeks = months_to_goal*4

print("Age:", my_age)
print("Study hours/week:", study_hours_per_week)
print("Months for goal:", months_to_goal)
print("Total number of weeks:", total_weeks)
print("typs:", type(my_age))
```

```python
# FLOATS - Numbers with decimal points
# Used for percentages, prices, measurements

course_fee = 9999.99
completion_percentage = 8.5
average_daily_hours = 1.5
total_days = 180
total_learning_hours = average_daily_hours*total_days

print("Course fee:", course_fee)
print("Completion %:", completion_percentage)
print("Average hours/day:", average_daily_hours)
print("Total days:", total_days)
print("Total learning hours:", total_learning_hours)
print("type:", type(course_fee))
```

```python
# BOOLEAN - only two posisble values : TRUE or FALSE
# Used for yes/no decission manking in your code

is_course_AI = True
is_course_completed = False
has_python_installed = True
has_completed_the_phase1 = False

print("The course name is AI:", is_course_AI)
print("Course completed:", is_course_completed)
print("Python installed:", has_python_installed)
print("Phase1 completed:", has_completed_the_phase1)
print("type:", type(is_course_AI))
```
